```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
```

```python
saldf=pd.read_csv('/content/drive/MyDrive/extra_need/Datas/Salary Data.csv')
```

```python
saldf.head()
```

```python
saldf.info()
```

```python
saldf.isnull().sum()
```

```python
saldf.dropna(inplace=True, how='all')
```

```python
saldf.isnull().sum()
```

```python
inp=saldf[['Years of Experience']]
out=saldf['Salary']
```

```python
LR=LinearRegression()
```

```python
LR.fit(inp,out)
```

```python
Years =float(input("Enter year of experience:"))
predicted_salary = LR.predict([[Years]])
print("Predicted salary:", predicted_salary[0],"\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Years of Experience', y='Salary', data=saldf)
plt.plot(saldf['Years of Experience'], LR.predict(saldf[['Years of Experience']]), color='red')
plt.title('Salary vs. Years of Experience with Regression Line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
