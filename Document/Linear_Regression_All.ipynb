Sales Prediction (Linear Regression) (BY:P SWATHIKA)

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
ad=pd.read_csv('/content/drive/MyDrive/extra_need/Datas/Advertising.csv')
ad.head(10)
```

```python
ad.info()
```

```python
ad.isnull().sum()
```

```python
ad.dropna(inplace=True,how='all')
inp=ad[['TV']]
out=ad['sales']
LR=LinearRegression()
LR.fit(inp,out)
```

```python
pred=LR.predict([[240]])
print(pred)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a scatter plot
sns.scatterplot(x=ad['TV'], y=ad['sales'], label='Actual Sales')

# Predict sales values using the trained LR model
predicted_sales = LR.predict(ad[['TV']])

# Overlay the predicted sales values as a line
plt.plot(ad['TV'], predicted_sales, color='red', label='Linear Regression Line')

# Add title and labels
plt.title('TV Advertising Spend vs. Sales with Linear Regression Line')
plt.xlabel('TV Advertising Spend')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
