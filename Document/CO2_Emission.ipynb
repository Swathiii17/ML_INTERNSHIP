### ***CO2 Emission Calculator using Multi Linear Regression by Vishwa.R***

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import pickle
```

```python
stmh=pd.read_csv('/content/drive/MyDrive/Data-Sets/CO2 Emissions.csv')
```

```python
stmh.head(10)
```

```python
stmh.info()
```

```python
stmh.isnull().sum()
```

```python
inp=stmh[['Engine Size(L)','Fuel Consumption Hwy (L/100 km)',]]
out=stmh['CO2 Emissions(g/km)']
```

```python
LR=LinearRegression()
```

```python
LR.fit(inp,out)
```

```python
filename = 'CO2.sav'
pickle.dump(LR, open(filename, 'wb'))
```

```python
pred=LR.predict([[2.4,11.2]])
print(pred)
```
